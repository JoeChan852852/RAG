# 🔌 Lesson: What is an API and Why Use One?

## 1. What is an API?

    An API (Application Programming Interface) is a set of rules that allows one piece of software to talk to another.

        Imagine you are at a restaurant:

            You are the User (Client).

            The Kitchen is the System (Server/Azure).

            The Waiter is the API.

    You don't go into the kitchen and start cooking; you give your order to the waiter, the waiter tells the kitchen, 
    and then the waiter brings your food back.

## 2. Why do we use them?

    We use APIs because they provide Abstraction and Efficiency:

    Benefits
        -Don't Reinvent the Wheel,"Instead of building your own AI, you use the Azure OpenAI API to ""borrow"" theirs."
        -Security,You don't give users access to your whole database; you give them an API that only shows what they need.
        -Standardization,It doesn't matter if your app is on iPhone or Android; both can talk to the same Azure API.

## 3. How do you use an API?

    Most modern APIs (like Azure's) use REST (Representational State Transfer). To use one, you need four things:

        The URL (Endpoint): Where the "waiter" is standing.

        The Method: What are you doing? (GET data, POST new data, DELETE data).

        Headers: Your ID card (API Key).

        Body: The actual information you are sending (usually in JSON format).

## 6.What are tokens

    Token is about How much you use.

        What it is: A small chunk of text (roughly 4 characters or 0.75 of a word).

        Why we use it: Azure AI doesn't charge you per "request"—it charges you based on how much text 
        the AI had to "read" and "write."

        Analogous to: The electricity meter in your house. You aren't charged for "turning on the lights"; 
        you are charged for the kilowatt-hours you consume.

### 📊 Token vs. Charge: How it works in Azure

    In Azure OpenAI or Cognitive Services, your bill is calculated using this formula:

$$ \Large \text{Total Cost} = ( \text{Input Tokens} \times \text{Price per Token})+ (\text{Output Tokens} \times \text{Price per Token}) $$

```python
def simulate_billing(text):
    # Rough rule: 1 token ~= 4 characters
    token_count = len(text) // 4
    price_per_1k_tokens = 0.002 # Example price ($0.002)
    
    total_cost = (token_count / 1000) * price_per_1k_tokens
    
    return {
        "text": text,
        "estimated_tokens": token_count,
        "estimated_charge": f"${total_cost:.6f}"
    }

# Example usage
user_prompt = "Explain how gravity works on Mars compared to Earth."
usage = simulate_billing(user_prompt)

print(f"Prompt: {usage['text']}")
print(f"Tokens Used: {usage['estimated_tokens']}")
print(f"Actual Charge: {usage['estimated_charge']}")
```

### 💡 Pro-Tip:

    Input Tokens (Prompts): These are usually cheaper.

    Output Tokens (Responses): These are usually more expensive because the AI has to "think" more to generate them.
