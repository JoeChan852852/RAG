# What is n8n? 🤔


    Imagine you wrote a Python script to check your email, download an attachment, and save it to a specific folder. 
    That’s great, but you have to host it, deal with API updates, manage passwords, and handle errors when things crash.

    n8n is a visual workflow automation tool that does all of that for you. It allows you to connect different apps and 
    services together using a drag-and-drop interface.

    Think of it like a flowchart from your first-year logic design class, but instead of just mapping out ideas, the 
    flowchart actually executes the code.

    In n8n, you pull visual blocks (called nodes) onto a canvas and connect them with wires. Data flows through these 
    wires from one app to another. It is free, open-source (technically "fair-code"), and self-hostable. While it is a 
    "low-code" platform, it’s built with developers in mind. If the visual blocks aren't enough, you can drop in a "Code" 
    node and write custom JavaScript to manipulate your data exactly how you want.

# Why should engineering students care? 🛠️

    Why use n8n instead of just writing a Python or C script? Because n8n handles the annoying parts of 
    programming (authentication, API limits, retries) so you can focus on the results.

    Here are a few realistic ways you can use n8n in your student life:

        GitHub Group Project Alerts: Working on a software engineering group project? Send an automatic Discord or Slack 
        message whenever a teammate pushes new code or creates a pull request.

        Hardware Data Logging: Have an Arduino or Raspberry Pi reading temperature sensors for a mechatronics lab? Have 
        your microcontroller send a simple web request to n8n, which then nicely formats and logs the data into a Google 
        Sheet in real-time.

        Course Announcements Scraper: Automatically check your university's RSS feed or course portal and text you if a 
        new assignment is posted.

        Simple AI Experiments: Connect OpenAI's API to summarize long, dense lecture PDFs and send the bullet points 
        straight to your Notion workspace.

        Daily Weather & Transit Briefing: Every morning at 7:30 AM, fetch the weather and the local bus schedule, and 
        send it to your Telegram so you know exactly when to leave for your 8 AM lecture.

# Your Very First Workflow: The Morning Briefing ☀️

    Let’s build something instantly useful. We are going to build a workflow that runs every day at 8 AM, 
    checks the weather, and sends you a message on Discord.

## Prerequisites

    Before you start in n8n, you need a Webhook URL from your Discord server:

        Open Discord, go to the channel where you want the weather updates, and click the gear icon (Edit Channel).

        Go to Integrations > Webhooks > New Webhook.

        Name it (e.g., "HK Weather Bot") and click Copy Webhook URL. Save this for later.

## Step 1: Set Up the Schedule Trigger

    This node will act as your alarm clock to run the workflow at 8:00 AM every day.

        Add a new node and search for Schedule Trigger.

        Under Trigger Interval, set it to Days.

        Set Days Between Runs to 1.

        Set Time of Day to 8:00 AM.

        Important Note: Make sure your n8n environment timezone is set to Hong Kong (Asia/Hong_Kong). 
        You can check this in n8n under Settings > Workflow Settings.

## Step 2: Fetch the Weather Data (Hong Kong)

    We will use an HTTP Request node to pull data from Open-Meteo.

        Click the + button next to your Schedule node and add an HTTP Request node.

        Method: GET

        URL: Paste the following URL (this asks for the current temperature, humidity, and weather conditions in Hong Kong):
            (https://api.open-meteo.com/v1/forecast?latitude=22.3193&longitude=114.1694&current=temperature_2m,relative_humidity_2m,wind_speed_10m&timezone=Asia%2FHong_Kong)

    Click Execute Node. You should see a JSON output with the current weather data!

## Step 3: Add News to Flow

    Add the RSS Read Node

        Insert a new node between your Weather node and your Gemini node.

        Search for RSS Read.

        URL: Use this official Hong Kong news feed:
        https://www.news.gov.hk/en/common/html/topstories.rss.xml

    Click Execute Node. It will return a list of the latest stories. 
    By default, n8n will pass only the most recent one to the next node.

## Step 4:Add AI to Flow

    The easiest way (using the dedicated node):

    Add the OpenRouter node.
    
    Prompt:
        You are a friendly Hong Kong local assistant.

        Weather Data: {{ $node["Get HK Weather"].json.current.temperature_2m }}°C, {{ $node["Get HK Weather"].
        json.current.relative_humidity_2m }}% humidity.
    
        Latest News: {{ $json.title }}

        Write a 3-sentence morning briefing for Discord.

        Start with a weather greeting and a practical tip (e.g., umbrella or clothing).

        Mention the news headline briefly in a conversational way ("In other news, I saw that...").

        End with a positive "Have a great day in HK!" message. Use emojis and a friendly tone.

## Step 5: Send the Discord Message

    Now, let's format that data and push it to Discord.

        Click the + button next to the LLM node and add a Discord node.

        Authentication: Select Webhook URL.

        Webhook URL: Paste the URL you copied from Discord earlier.

        Resource: Message

        Operation: Send

        Content: {{ $json.text }}
